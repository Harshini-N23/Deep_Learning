```python
import cv2
import cvlib as cv
from google.colab.patches import cv2_imshow
from cvlib.object_detection import draw_bbox
import numpy as np
```

```python
image = cv2.imread("/content/OD.jpg")
bbox, label, conf = cv.detect_common_objects(image)
output_image = draw_bbox(image, bbox, label, conf)
cv2_imshow(output_image)
```

```python
image = cv2.imread("/content/od4.jfif")
# apply face detection
faces, confidences = cv.detect_face(image)
print("faces: ",faces)
print("Confidences: ",confidences)

for face,conf in zip(faces,confidences):
    (startX,startY) = face[0],face[1]
    (endX,endY) = face[2],face[3]
    cv2.rectangle(image, (startX,startY), (endX,endY), (0,255,0), 2)
cv2_imshow(image)
```

## Conclusion:
### Using the Computer Vision Library function, ccv.detect_common_objects and cv.detect_face we were able to identify the objects. The identified objects are highlighted by bounding boxes.
